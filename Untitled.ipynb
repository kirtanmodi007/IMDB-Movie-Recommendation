How to work with missing data?</h3>

Steps for working with missing data:

<b>1. Identify missing data<br>
<b>2. Deal with missing data<br>
<b>3. Correct data format<br>


<h3>STEP - 1 Identify Missing Values 

```python
import numpy as np
```

```python
#We are converting all the "?" in our dataset to NaN (Not A Number) format.
for each_column in df:
    print(each_column)
    #Now we are replacing all "?" with "NaN".
    df.replace("?", np.nan,inplace = True)
```

```python
df
```

```python
#Now we have to check how many Null values do we have in each column.
#value_counts() function return all the unique values and their frequency in particular column.
#True means null (NaN) value, False means that particular cell has some value.
for each_column in df:
    print(each_column)
    print(df[each_column].isna().value_counts())
    print("")
```

Based on the summary above, each column has 205 rows of data, seven columns containing missing data:
<ol>
    <li>"normalized-losses": 41 missing data</li>
    <li>"num-of-doors": 2 missing data</li>
    <li>"bore": 4 missing data</li>
    <li>"stroke" : 4 missing data</li>
    <li>"horsepower": 2 missing data</li>
    <li>"peak-rpm": 2 missing data</li>
    <li>"price": 4 missing data</li>
</ol>

<p> One thing we are clear is that, either of the car will have 2 doors or 4 doors.
<p> So, for the missing value in "num-of-doors" column we can put 4 as it is the most frequent value. 

```python
#So, whenever we see "NaN" in "num-of-doors" column we just have to replace it with "four".
df['num-of-doors'].replace(np.nan,"four",inplace = True)
```

```python
#SO, we can see that in "num-of-doors" column every row has either "two" or "four" as its value. No null value.
df['num-of-doors'].isna().value_counts()
```

```python
#Checking all the unique value in 'num-of-doors' column.
df['num-of-doors'].value_counts()
```

<p> As we can see 'num-of-doors' column has value in string format. But having this column in integer format makes more sense.
Because it makes further data analysis process smooth.

```python
#Replacing "four" with 4(Integer) and "two" with 2(Integer) in "num-of-doors" column.
df['num-of-doors'].replace("four",4,inplace = True)
df['num-of-doors'].replace("two",2,inplace = True)
```

```python
df['num-of-doors'].value_counts()
```

```python
#Now if we check the type of "num-of-doors" column we can see that it is in Integer format.
df['num-of-doors'].dtype
```

<h5> Now we are cleaning / formating 'bore' column.

```python
#Firstly check how many null values do we have in 'bore' column.
#We can see currently there are 4 null values.
#df['bore'].isna().value_counts()
df['bore'].isna().value_counts()
```

<h4>isnull() and isna() both functions give same result.

```python
#Printing only those rows where 'bore' colum has NaN value.
df[df['bore'].isna()]
```

<h4> Now we are replacing the NaN values with mean values of that column.
<h4> But as we know 'bore' column contains value in string format, so we have to change it to Float.

```python
df['bore'] = df['bore'].astype('float64')
```

```python
df['bore'].describe()
```

<h4> Now in 'bore' column we have to convert NaN with MEAN() values

```python
meanValue = df['bore'].mean()
df['bore'].replace(np.nan, meanValue,inplace = True)
```

```python
#This simply means no row contains null value for 'bore' column.
df['bore'].isna().value_counts()
```

<h4> Now we are cleaning / formating 'stroke' column.
    <p> As we can see 'stroke' column is of Object type but it conatins numeric values. So, we have to convert it into float64.

```python
df['stroke'].dtypes
```

```python
df['stroke'] = df['stroke'].astype('float64')
```

```python
#So, as our 'stroke' column contains float value, now we just have to replace nan values with mean values.
#Printing 'nan' value rows

df[df['stroke'].isna()]
```

<h4> Now replace all the NaN with mean values.

```python
meanValue = df['stroke'].mean()
df['stroke'].replace(np.nan, meanValue, inplace = True)
```

<h4>By running this code we can see that now we don't have any NaN value in stroke column.


```python
df[df['stroke'].isna()]
```

```python
for each_column in df:
    print(each_column)
    print(df[each_column].isna().value_counts())
    print("")
```

<h4>Drop the whole row:</h4><br>
<li>"price": 4 missing data, simply delete the whole row
<li>Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us</li>
