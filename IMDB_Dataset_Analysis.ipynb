```python
dictionary_For_Count_Occurance = {}
for particular_Genre in df['genres']:
    for one_genre in particular_Genre:
        if one_genre in dictionary_For_Count_Occurance:
            dictionary_For_Count_Occurance[one_genre] = dictionary_For_Count_Occurance[one_genre] + 1
        else:
            dictionary_For_Count_Occurance[one_genre] = 1
```

```python
dictionary_For_Count_Occurance
```

```python
#Making an Empty List to Store all the keys.
keyList = []
for key in dictionary_For_Count_Occurance.keys():
    keyList.append(key)

print(keyList)

#Making an empty List to store all the values.
valueList = []
for value in dictionary_For_Count_Occurance.values():
    valueList.append(value)

print(valueList)
    
#So, now we have our genres and its count in 2 different LISTS. Which we can use for plotting purpose.
```

```python
#Importing libraries for Plotting Purpose
import matplotlib.pyplot as plt

plt.barh(keyList, valueList, color = "#0000ff")
plt.figure(figsize=(50,20))
plt.show()
```
